```python
import shutil # save img locally
```

```python
url = 'https://imgs.xkcd.com/comics/sunspot_cycle_2x.png'
```

```python
file_name = "LAB_13_picture.png"
```

```python
res = requests.get(url, stream = True)
```

```python
if res.status_code == 200:
    with open(file_name,'wb') as f:
        shutil.copyfileobj(res.raw, f)
    print('Image sucessfully Downloaded: ',file_name)
else:
    print('Image Couldn\'t be retrieved')
```

```python
#Using Python, call Github API to find out the cold joke contained in the 24 secret files in the following repo:

#https://github.com/ironhack-datalabs/scavenger

#The filenames of the secret files contain .scavengerhunt and they are scattered in different 
#directories of this repo. The secret files are named from .0001.scavengerhunt to .0024.scavengerhunt. 
#They are scattered randomly throughout this repo. You need to search for these files by calling the Github API,
#not searching the local files on your computer.

#Notes:
## https://docs.github.com/en/rest/search?apiVersion=2022-11-28#about-search

response = requests.get('https://api.github.com/search/code?q=.scavengerhunt+in:path+repo:ironhack-datalabs/scavenger')
#api.github - is the link where are all the apis. the .scavengerhunt is an extension , we search in code -> in the path name of the file
# + is an extra condition , + repo (repository) " ironhack-datalabs"
search = response.json()
search
```

```python
search["items"]
```

```python
repo_files_names = [item["name"] for item in search["items"]]
repo_files_names.sort()
repo_files_names
```

```python
repo_files_urls = {item["name"]: item["url"] for item in search["items"]}
repo_files_urls
```

```python
repo_files = [requests.get(repo_files_urls[repo_filename]) for repo_filename in repo_files_names]
repo_files
```

```python
repo_files_content = [file.json()["content"] for file in repo_files]
repo_files_content
```

```python
import base64
```

```python
message = [base64.b64decode(b64message).decode('ascii') for b64message in repo_files_content]
message
```
